```python
def gauss_function(x, a, x0, sigma):
    return a*np.exp(-(x-x0)**2/(2*sigma**2))


def compute_rms(spec_axis, spectrum, ContLower1, ContUpper1, ContLower2, ContUpper2):

    cont_channels = np.where((spec_axis < ContUpper1) | (spec_axis > ContLower2))

    continuum_vals = spectrum[cont_channels]
    

    # calculate the root mean square of the continuum
    # this will be our uncertainty
    rms = np.sqrt(np.mean(np.square(continuum_vals)))
    
    return rms



def red_chisq(obs, calc, err, free_params):

    """
    Function to calculate a reduced chi-square.

    Parameters
    -------------------

    obs : array
        Array of observed values.

    calc : array
        Array of calculated values (i.e., a model).

    err : array, int, or float
        Err

    sigma : int or float
        Width of the Gaussian.

    """

    return((np.sum((obs-calc)**2 / err**2))/(len(obs)-free_params))


def make_slice(rect, cube, outflow_blue, ERRoutflow_blue, width, length, pa, CO_rot_point):
       
    # extract values along the slice
    [xlowL, ylowL], [xlowR, ylowR], [xupR, yupR], [xupL, yupL] = rect.get_corners()
    xLine = np.linspace(int(round((xlowL+xlowR)/2)), int(round((xupL+xupR)/2)), round(length))
    yLine = np.linspace(int(round((ylowL+ylowR)/2)), int(round((yupL+yupR)/2)), round(length))
    blue_vels = outflow_blue[yLine.astype(np.int), xLine.astype(np.int)]
    ERRblue_vels = ERRoutflow_blue[yLine.astype(np.int), xLine.astype(np.int)]
    
    # find the separation between each pixel in the slice
    origin = SkyCoord.from_pixel(xupL, yupL, wcs=CO10w)  # top of the slice
    seps = []
    for i in range(len(xLine)):  # distance from origin, starting at bottom of slice
        next_pix = SkyCoord.from_pixel(xLine[i],yLine[i], wcs=CO10w)
        seps.append(origin.separation(next_pix).arcsec)
    seps = np.array(seps)
    
    # slice the CO cube
    COrect = Rectangle((CO_rot_point[0],CO_rot_point[1]-length//2), width=width, height=length, angle=pa, rotation_point=CO_rot_point)
    [xlowL, ylowL], [xlowR, ylowR], [xupR, yupR], [xupL, yupL] = COrect.get_corners()
    xLine = np.linspace(int(round((xlowL+xlowR)/2)), int(round((xupL+xupR)/2)), round(length))
    yLine = np.linspace(int(round((ylowL+ylowR)/2)), int(round((yupL+yupR)/2)), round(length))
    
    # make a mask
#     mask = np.zeros((cube[0].shape[1], cube[0].shape[2]))
#     mask[yLine.astype(np.int), xLine.astype(np.int)] = 1.
    
#     print(yLine.astype(np.int), xLine.astype(np.int))
    
    return blue_vels, ERRblue_vels, xLine, yLine, seps


def guesses(spec, i, j, which_CO, which_slice):
    
    if (which_CO == 'CO21') & (which_slice == 'SW'):
        if (i == 226) & ((j==171) | (j==172)):
            amp1 = 0.06
            vel1 = 275
            sig1 = 20
            amp2 = 0.1
            vel2 = 410
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                     limits=[(0,0), (200,300), (0,0), (0,0), (350,450), (0,0)], 
                     limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
                     annotate = False, quiet=True, negamp=False)    
        else:
            amp1 = 0.06
            vel1 = 200
            sig1 = 20
            amp2 = 0.1
            vel2 = 400
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                 annotate = False, quiet=True, negamp=False)
            
            
    elif (which_CO == 'CO21') & (which_slice == 'SE'):

        # initial guesses
#         if (164 <= i <= 173) & (189 <= j <= 196):
#             amp1 = 0.06
#             vel1 = 250
#             sig1 = 20
#             amp2 = 0.1
#             vel2 = 310
#             sig2 = 20
#             spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
#                      limits=[(0,0), (200,290), (0,0), (0,0), (290,320), (0,0)], 
#                      limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
#                      annotate = False, quiet=True, negamp=False)
            
        if (174 <= i <= 182) & (197 <= j <= 203):
            amp1 = 0.06
            vel1 = 195
            sig1 = 20
            amp2 = 0.1
            vel2 = 295
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                     limits=[(0,0), (150,210), (0,0), (0,0), (250,320), (0,0)], 
                     limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
                     annotate = False, quiet=True, negamp=False)  
            
        else:
            amp1 = 0.06
            vel1 = 200
            sig1 = 20
            amp2 = 0.1
            vel2 = 300
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                                  limits=[(0,0), (100,350), (0,0), (0,0), (100,350), (0,0)], 
                                  limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)], 
                                  annotate = False, quiet=True, negamp=False)
            
    return spec


def fit_CO(savepath, cube_file, xLine, yLine, which_CO, which_slice):
    
    if which_CO == 'CO21':
        restfreq = 230.538
    
    cube = SpectralCube.read(cube_file, hdu=0)
    subcube = cube.with_spectral_unit(u.km / u.s, velocity_convention='radio', rest_value=restfreq * u.GHz)

    z, y, x = cube.shape

    done_pixel_list = []
#     pixel_list = set(list(zip(xLine.astype(int), yLine.astype(int))))
    
    count = 0
    for pixel in pixel_list:
    
        i, j = pixel[0], pixel[1]

        # adding a check to avoid duplicates due to rounding of pixel number
        if (i,j) in done_pixel_list:
            count+=1
            print('Pixel %s/%s' % (count, len(xLine)), end='\r')
            continue
        else:
            done_pixel_list.append((i,j))

        # grab the spectrum and do the fit
        spectrum = np.array(subcube[:,j,i], dtype='float64') 
        minval = 0
        maxval = max(np.array(subcube.spectral_axis))
        spec_axis = np.linspace(minval, maxval, len(spectrum))

        if (np.isfinite(np.mean(spectrum)) == False): #| (mask[j,i] == 0.):
            count+=1
            print('Pixel %s/%s' % (count, len(xLine)), end='\r') 
            continue

        spec = pyspeckit.Spectrum(data=spectrum, xarr=spec_axis)

        # grab the proper guesses
        spec = guesses(spec, i, j, which_CO=which_CO, which_slice=which_slice)

        # get the parameters and the errors
        spec.measure()
        ampERR1, velERR1, sigERR1, ampERR2, velERR2, sigERR2 = spec.specfit.parinfo.errors
        amps_list = []
        centers_list = []
        widths_list = []
        for line in spec.measurements.lines.keys():
            amps_list.append(spec.measurements.lines[line]['amp'])
            centers_list.append(spec.measurements.lines[line]['pos'])
            widths_list.append(spec.measurements.lines[line]['fwhm']/2.355)
            params = [par for sublist in zip(amps_list, centers_list, widths_list) for par in sublist]


        # calculate reduced chi square      
        if (which_CO == 'CO21'):
            err_redchisq = compute_rms(spec_axis, spectrum, 0, 100, 450, 540)
            
        # now get the rms of the residuals
        rms_resids = round(np.sqrt(np.mean(np.square(np.array(spec.specfit.residuals)))),3)

        components = [gauss_function(np.linspace(minval, maxval, len(spectrum)), 
                        amps_list[i], centers_list[i], widths_list[i]) 
                        for i in np.arange(len(amps_list))]
        model = sum(components)

        redchisq = round(red_chisq(spectrum, model, err=err_redchisq, free_params=6), 4)

        with open("%s/fits.txt" % savepath, "a") as f:
                f.write('%s, %s, %s, %s, %s, %s, %s, %s, %s\n' % (i, j, round(redchisq,2), params[0], params[1], params[2], 
                                                                  params[3], params[4], params[5]))
                f.close()

        with open("%s/fits_err.txt" % savepath, "a") as e:
                e.write('%s, %s, %s, %s, %s, %s, %s, %s, %s, %s\n' % (i, j, err_redchisq, rms_resids, 
                                                                      ampERR1, velERR1, sigERR1, 
                                                                      ampERR2, velERR2, sigERR2))
                e.close()

        # plot the spectrum and the fit
        spec.plotter()
        ax = plt.subplot(1,1,1)
        plt.rcParams["figure.figsize"] = (10,5)
        spec.specfit.plot_fit(annotate=False, 
                              show_components=False,
                              composite_fit_color='tab:pink',
                              lw=1.5)

        spec.specfit.plot_components(component_fit_color='tab:cyan',
                                    lw=1.5)

        spec.specfit.plotresiduals(axis=spec.plotter.axis,
                            clear=False,
                            yoffset=-0.2*np.max(spectrum), 
                            color='tab:purple',
                            linewidth=1.5)

        plt.xlabel('km/s [Vsys = 243 km/s]')
        plt.ylabel('Jy/beam')
        plt.ylim(bottom=-0.8*np.max(spectrum))

        at = AnchoredText('REDCHISQ: %s, AMP1: %s, VEL1: %s, SIGMA1: %s\n AMP2: %s, VEL2: %s, SIGMA2: %s' % 
                          (round(redchisq,2), round(params[0],2), round(params[1],2), round(params[2],2),
                          round(params[3],2), round(params[4],2), round(params[5],2)), frameon=False, loc='upper right')
        at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
        ax.add_artist(at)
        plt.title('Pixel: %s,%s' % (i,j))
        plt.savefig('%s/pixel_%s_%s.png' % (savepath, i,j), dpi=200)
        plt.close()

        count+=1
        print('Pixel %s/%s' % (count, len(pixel_list)), end='\r')

    return

# grab the velocities from the text file
def get_vels(savepath):
    
    CO = pd.read_csv('%s/fits.txt' % savepath)
    CO_err = pd.read_csv('%s/fits_err.txt' % savepath)
    
    CO = CO.sort_values(by=['X','Y'])
    CO_err = CO_err.sort_values(by=['X','Y'])

    CO['Vel1_Err'] = CO_err['Vel1']
    CO['Vel2_Err'] = CO_err['Vel2']
    CO['rms_resids'] = CO_err['rms_resids']

    CO_vels = []
    CO_errs = []
    for index, row in CO.iterrows():

        vel1 = row['Vel1']
        vel2 = row['Vel2']
        ERRvel1 = row['Vel1_Err'] * row['rms_resids']
        ERRvel2 = row['Vel2_Err'] * row['rms_resids']

        if vel1 < vel2:
            CO_vels.append(vel1)
            CO_errs.append(ERRvel1)
        elif vel2 < vel1:
            CO_vels.append(vel2)
            CO_errs.append(ERRvel2)
            
    return CO_vels, CO_errs
```
