```python
def guesses(spec, i, j, which_CO, which_slice):
    
    if (which_CO == 'CO21') & (which_slice == 'SW'):
        if (i == 226) & ((j==171) | (j==172)):
            amp1 = 0.06
            vel1 = 275
            sig1 = 20
            amp2 = 0.1
            vel2 = 410
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                     limits=[(0,0), (200,300), (0,0), (0,0), (350,450), (0,0)], 
                     limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
                     annotate = False, quiet=True, negamp=False)    
        else:
            amp1 = 0.06
            vel1 = 200
            sig1 = 20
            amp2 = 0.1
            vel2 = 400
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                 annotate = False, quiet=True, negamp=False)
            
            
    elif (which_CO == 'CO21') & (which_slice == 'SE'):

        # initial guesses
#         if (164 <= i <= 173) & (189 <= j <= 196):
#             amp1 = 0.06
#             vel1 = 250
#             sig1 = 20
#             amp2 = 0.1
#             vel2 = 310
#             sig2 = 20
#             spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
#                      limits=[(0,0), (200,290), (0,0), (0,0), (290,320), (0,0)], 
#                      limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
#                      annotate = False, quiet=True, negamp=False)
            
        if (174 <= i <= 182) & (197 <= j <= 203):
            amp1 = 0.06
            vel1 = 195
            sig1 = 20
            amp2 = 0.1
            vel2 = 295
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                     limits=[(0,0), (150,210), (0,0), (0,0), (250,320), (0,0)], 
                     limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)],
                     annotate = False, quiet=True, negamp=False)  
            
        else:
            amp1 = 0.06
            vel1 = 200
            sig1 = 20
            amp2 = 0.1
            vel2 = 300
            sig2 = 20
            spec.specfit.multifit(fittype='gaussian', guesses = [amp1, vel1, sig1, amp2, vel2, sig2], 
                                  limits=[(0,0), (100,350), (0,0), (0,0), (100,350), (0,0)], 
                                  limited=[(False, False), (True, True), (False, False), (False, False), (True, True), (False, False)], 
                                  annotate = False, quiet=True, negamp=False)
            
    return spec
```
